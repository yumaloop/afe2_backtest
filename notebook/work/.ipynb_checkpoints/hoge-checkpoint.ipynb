```python
import pandas_datareader
import pandas as pd
import yfinance as yf

from pprint import pprint

import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('seaborn') # style.use('ggplot') # style.use('fivethirtyeight')
%matplotlib inline

pd.set_option('display.max_columns', 500)
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
def get_df(symbol, st="2011/10/31", ed="2020/10/31"):
    df = pandas_datareader.data.DataReader(symbol, 'yahoo', st, ed) # daily
    df = df.sort_values(by='Date', ascending=True)
    df = df.resample('M').mean() # daily -> monthly
    df = df.fillna(method='ffill') # 1つ前の行の値で埋める
    df = df[['Close']].rename(columns={'Close': symbol})
    return df
```

```python
symbol = "^N225"
df = get_df(symbol, st="2000/1/1")

plt.figure(figsize=(12,6))
plt.plot(df.index.values, df[symbol].values)
plt.show()
```

```python
symbol = "BTC-USD"
df = get_df(symbol, st="2015/1/1")

plt.figure(figsize=(12,6))
plt.plot(df.index.values, df[symbol].values)
plt.show()
```

```python
stocks_topix30 = [2914, 3382, 4063, 4452, 4502, 
                  4503, 5401, 6301, 6501, 6502, 
                  6752, 6758, 6954, 7201, 7203, 
                  7267, 7751, 8031, 8058, 8306, 
                  8316, 8411, 8604, 8766, 8802, 
                  9021, 9432, 9433, 9437, 9984]

symbols = [str(s)+'.T' for s in stocks_topix30]
symbols
```

```python
for idx, symbol in enumerate(symbols):
    ticker = yf.Ticker(symbol)
    d = {k: pd.Series(v, dtype='object') for k, v in ticker.info.items()}
    df_one = pd.DataFrame(d)
    df = df_one if idx == 0 else df.append(df_one, ignore_index=True)
```
