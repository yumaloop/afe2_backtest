```python
import pandas_datareader
import pandas as pd
import yfinance as yf

from pprint import pprint

import matplotlib.pyplot as plt
%matplotlib inline

pd.set_option('display.max_columns', 500)
```

```python
stocks_topix30 = [2914, 3382, 4063, 4452, 4502, 
                  4503, 5401, 6301, 6501, 6502, 
                  6752, 6758, 6954, 7201, 7203, 
                  7267, 7751, 8031, 8058, 8306, 
                  8316, 8411, 8604, 8766, 8802, 
                  9021, 9432, 9433, 9437, 9984]

symbols = [str(s)+'.T' for s in stocks_topix30]
symbols
```

```python
for idx, symbol in enumerate(symbols):
    ticker = yf.Ticker(symbol)
    d = {k: pd.Series(v, dtype='object') for k, v in ticker.info.items()}
    df_one = pd.DataFrame(d)
    df = df_one if idx == 0 else df.append(df_one, ignore_index=True)
```
